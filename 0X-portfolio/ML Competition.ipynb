### Group 1

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**3 - 3*x**2 + 2*x

# Define the analytical derivative
def f_prime(x):
    return 3*x**2 - 6*x + 2

# Generate x values
x = np.linspace(-1, 3, 200)

# Compute function values
y = f(x)

# Compute derivative values
y_prime = f_prime(x)

# Plot the function and its derivative
plt.figure(figsize=(8,5))
plt.plot(x, y, label='f(x)')
plt.plot(x, y_prime, label="f'(x)", linestyle='--')
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)
plt.title("Function and its Derivative")
plt.xlabel("x")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
````

The **standard deviation** of a dataset measures how spread out the values are from the mean.  

For a dataset of $n$ values $x_1, x_2, \dots, x_n$ with mean $\bar{x}$, the **population standard deviation** is:

$$
\sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2}
$$
